```python
from collections.abc import Iterable
from numpy import *
import math
import matplotlib.pyplot as plt

def obtener_tipo(num):
    if isinstance(num, Iterable):
        ls = []
        for n in num:
            r = 3*n+2
            s = r*(3*n+2)+3
            S = gap(f'NumericalSemigroup({s},{s+3},{s+3*n+1},{s+3*n+2})')
            ls.append(int(gap(f'Type({S})')))
        return ls
    else:
        n = int(num)
        r = 3*n+2
        s = r*(3*n+2)+3
        S = gap(f'NumericalSemigroup({s},{s+3},{s+3*n+1},{s+3*n+2})')
        return int(gap(f'Type({S})'))
```

```python
obtener_tipo(arange(2, 20, 1))
```

```python
obtener_tipo(2)
```

```python
t = arange(2, 20, 1)

tipo_a = 2*t+3
tipo_b = asarray(obtener_tipo(t))

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(t,tipo_a,t,tipo_b)
plt.xticks(t)
ax.spines[['top', 'right']].set_visible(False)
ax.legend(['2n+3', 't(S)'])
plt.savefig('output_6_0.png')
plt.show()
```
